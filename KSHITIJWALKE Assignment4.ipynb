```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
plt.hist(df['age'])
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(df['age'])
plt.title("Boxplot of Age")
plt.show()
```

```python
plt.hist(df['age'], bins=10)
plt.title("Histogram (Alternative to Density)")
plt.show()
```

```python
plt.scatter(df['age'], df['chol'])
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.title("Scatter Plot")
plt.show()
```

```python
corr = df.corr()

plt.imshow(corr)
plt.colorbar()
plt.title("Heatmap")

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.show()
```

```python
df.groupby('sex')['chol'].mean().plot(kind='bar')
plt.title("Bar Plot (Sex vs Cholesterol)")
plt.show()
```

```python
df['sex'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
df['age'].plot()
plt.title("Line Chart")
plt.show()
```
